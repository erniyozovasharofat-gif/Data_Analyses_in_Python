```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student_habits_performance.csv")
data.info()
```

```python
# eng katta va eng kichik yoshni topish

age = data["age"].agg({
    "eng katta yosh" : "max",
    "eng kichik yosh" : "min"
})

print(age)
```

```python
# plt.style.use("seaborn-v0_8")

plt.hist(data["age"], bins = 10, color = "skyblue", edgecolor = "black")
plt.title("Talabalarning yosh taqsimoti")
plt.xlabel("Talabalarning yoshi")
plt.ylabel("Talabalarning soni")
plt.show()

plt.savefig ("Yoshlar taqsimoti.png")
```

```python
# Talabalarning o`qish soatining imtihon natijalariga tasiri

plt. scatter(data["study_hours_per_day"], data["exam_score"], alpha=0.8, color="maroon")
plt.title("Study hours affect for exam scores")
plt.xlabel("Study hours")
plt.ylabel("Exam scores")
plt.show()
```

```python
#Attandance percentage affect for exam scores

plt.scatter(data["attendance_percentage"], data["exam_score"], alpha= 0.8, color = "green", marker = "x")
plt.title("Attaedance percentage affect for exam scores")
plt.xlabel("Attendance percentage")
plt.ylabel("Exam scores")
plt.show()

```

```python
#bitta ekranda korsatish
fig, axes = plt.subplots(1, 2 , figsize = (10 , 5))

# Talabalarning o`qish soatining imtihon natijalariga tasiri

axes[0]. scatter (data["study_hours_per_day"], data["exam_score"], alpha=0.8, color="maroon")
axes[0].set_title("Study hours affect for exam scores")
axes[0].set_xlabel("Study hours")
axes[0].set_ylabel("Exam scores")


#Attandance percentage affect for exam scores

axes[1].scatter(data["attendance_percentage"], data["exam_score"], alpha = 0.8, color = "green", marker = "x")
axes[1].set_title("Attedance percentage affect for exam scores")
axes[1].set_xlabel("Attendance percentage")
axes[1].set_ylabel("Exam scores")

plt.show()
```

```python
# Age affect for socialmedia hours and netflix hours

# Social Media scatter
x1 = np.array(data["social_media_hours"])
y1 = np.array(data["age"])
plt.scatter(x1, y1, color="skyblue", alpha=0.9, s=10, label="Social Media")

# Netflix scatter
x2 = np.array(data["netflix_hours"])
y2 = np.array(data["age"])
plt.scatter(x2, y2, color="red", alpha=0.9, s=10, label="Netflix")

# Labels and title
plt.xlabel("Hours Spent on Platform")
plt.ylabel("Age")
plt.title("Scatter Plot: Social Media & Netflix Hours vs Age")
plt.legend()
plt.show()
```
